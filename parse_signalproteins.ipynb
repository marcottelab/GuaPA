```python
import pandas as pd
import numpy as np
import subprocess
import sys
sys.path.append("..")
from src.useful import read3line_tmbed
from src.useful import fasta2df
```

```python
usp = pd.read_csv('PETases_uniquephyla_12152023_predictedSPs_results.csv', header=0)
usp.head()
```

```python
tmbed = read3line_tmbed('PETHits_uniquephyla_Guaymas2020_ALLBINS_12152023.fasta.tmbed.3lines')
tmbed.head()
```

```python
#Count number of instances of 'S' in each row of the TMBed_prediction column
tmbed['sig_prot'] = tmbed['TMBed_prediction'].str.count('S')
tmbed.head()
```

```python
#Read in the SP6 data
sp6 = pd.read_csv('SP6_PETHITs_prediction_results.txt', skiprows=1, header=0, sep='\t')
sp6.head()
```

```python
sp6_pro = fasta2df('SP6_PETHitsprocessed_entries.fasta.txt')
sp6_pro.head()
```

```python
#rename column
sp6_pro.rename(columns={'aa_sequence': 'sp6_aa'}, inplace=True)
```

```python
usp.head()
```

```python
def trim_prot_usp(row):
    snippet = row['predicted_cleavage']
    sequence = row['sequence']
    if snippet in sequence: 
        dropped = sequence.replace(snippet, '')
    return dropped 
```

```python
#subtract the predicted_cleavage column from the sequence column to get the protein without signal peptide
usp.dropna(subset=['predicted_cleavage'], inplace=True)
usp['usp_aa'] = usp.apply(lambda x: trim_prot_usp(x), axis=1)
usp.head()
```

```python
#FIrst need to create the suggested tmbed sequences
def trim_prot_tmbed(row):
    num_sig = row['TMBed_prediction'].count('S')
    sequence = row['aa_sequence']
    if num_sig == 0:
        return None
    else:
        return sequence[num_sig:]
```

```python
tmbed['tmbed_aa'] = tmbed.apply(lambda x: trim_prot_tmbed(x), axis=1)
tmbed.head()
```

```python
#Join on sp6_pro with usp and tmbed
tmbed_sp6 = tmbed.merge(sp6_pro, how='left', on='ProteinID')
tmbed_sp6_usp = tmbed_sp6.merge(usp, how='left', left_on='aa_sequence', right_on='sequence')
```

```python
tmbed_sp6_usp.head()
```

```python
tmbed_sp6_usp.drop(columns=['sequence', 'predicted_cleavage', 'TMBed_prediction', 'sig_prot', 'predicted_cleavage', 'predicted_type'], inplace=True)
```

```python
tmbed_sp6_usp.head()
```

```python
tmbed_sp6_usp.to_csv('PETases_uniquephyla_12152023_predictedSPs.csv', index=False)
```
